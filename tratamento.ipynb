```python
import pandas as pd

df_video = pd.read_csv('/content/videos-stats.csv')
df_comentario = pd.read_csv('/content/comments.csv')
```

```python
print("Schema para df_video:")
df_video.info()

print("\nSchema para df_comentario:")
df_comentario.info()
```

```python
df_video[['Likes', 'Comments', 'Views']] = df_video[['Likes', 'Comments', 'Views']].fillna(0)
df_comentario[['Likes', 'Sentiment']] = df_comentario[['Likes', 'Sentiment']].fillna(0)
```

```python
print(f"Numero de registros em df_video: {df_video.shape[0]}")
print(f"Numero de registros em df_comentario: {df_comentario.shape[0]}")
```

```python
df_video.dropna(subset=['Video ID'], inplace=True)
df_comentario.dropna(subset=['Video ID'], inplace=True)

print(f"Numero de registros restantes em df_video depois de remover nulos: {df_video.shape[0]}")
print(f"Numero de registros restantes em df_comentario depois de remover nulos: {df_comentario.shape[0]}")
```

```python
df_video.drop_duplicates(subset=['Video ID'], keep='first', inplace=True)
print(f"Numero de registros restantes em df_video depois de remover duplicatas: {df_video.shape[0]}")
```

```python
df_video['Likes'] = df_video['Likes'].astype(int)
df_video['Comments'] = df_video['Comments'].astype(int)
df_video['Views'] = df_video['Views'].astype(int)

df_comentario.rename(columns={'Likes': 'Likes Comment'}, inplace=True)
df_comentario['Likes Comment'] = df_comentario['Likes Comment'].astype(int)
df_comentario['Sentiment'] = df_comentario['Sentiment'].astype(int)

print("Tipos de dados após a conversão:")
print(df_video[['Likes', 'Comments', 'Views']].dtypes)

print("\nTipos de dados após a conversão:")
print(df_comentario[['Likes Comment', 'Sentiment']].dtypes)
```

```python
df_video['Interaction'] = df_video['Likes'] + df_video['Comments'] + df_video['Views']
display(df_video.head())
```

```python
df_video['Published At'] = pd.to_datetime(df_video['Published At'])
print(df_video.dtypes)
```

```python
df_video['Year'] = df_video['Published At'].dt.year
display(df_video.head())
```

```python
df_join_video_comments = pd.merge(df_video, df_comentario, on='Video ID', how='left')
display(df_join_video_comments.head())
```

```python
df_us_videos = pd.read_csv('/content/USvideos.csv', engine='python', quotechar='"', on_bad_lines='skip')
display(df_us_videos.head())
```

```python
print("Schema para df_us_videos:")
df_us_videos.info()
```

```python
df_join_video_usvideos = pd.merge(df_video, df_us_videos, left_on='Title', right_on='title', how='left')
display(df_join_video_usvideos.head())
```

```python
null_counts = df_video.isnull().sum()
display(null_counts)
```

```python
if '_c0' in df_video.columns:
    df_video.drop(columns=['_c0'], inplace=True)

df_video.to_parquet('videos-tratados-parquet', index=False)
```

```python
if '_c0' in df_join_video_comments.columns:
    df_join_video_comments.drop(columns=['_c0'], inplace=True)

df_join_video_comments.to_parquet('videos-comments-tratados-parquet', index=False)
```
